```python
# —Å –∫–∞–∂–¥–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã —Å–æ–±–µ—Ä–∏—Ç–µ —Å—Å—ã–ª–∫–∏ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω—ã–µ –Ω–æ–≤–æ—Å—Ç–∏
# –ø–æ–¥—Å–∫–∞–∑–∫–∞: —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏–π —Ç–µ–≥ h2 (–ø—Ä–æ–≤–µ—Ä—è–π—Ç–µ –ø–æ –Ω–µ–º—É), –Ω—É–∂–Ω—ã–π –Ω–∞–º - a

for i1 in soup1.find_all('a'):
  if i1.parent.name == 'h2':
    print(i1.get('href'))

for i2 in soup2.find_all('a'):
  if i2.parent.name == 'h2':
    print(i2.get('href'))

for i3 in soup3.find_all('a'):
  if i3.parent.name == 'h2':
    print(i3.get('href'))

# –≤–∞—à –∫–æ–¥

```

```python
# –¥–ª—è –ø–∞—Ä—Å–∏–Ω–≥–∞ –¥–∞—Ç—ã: –∏–∑–≤–ª–µ–∫–∏—Ç–µ —Ç–µ–∫—Å—Ç –∏–∑ —Ç–µ–≥–∞ time
link = 'https://sysblok.ru/education/defenestracija-ne-ne-slyshal/'
page = requests.get(link)
soup = BeautifulSoup(page.text, features="html.parser")
print(soup.find_all('time')[0].text)
```

```python
# –¥–ª—è –ø–∞—Ä—Å–∏–Ω–≥–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞: –∏–∑–≤–ª–µ–∫–∏—Ç–µ —Ç–µ–∫—Å—Ç –∏–∑ —Ç–µ–≥–∞ h1
print(soup.find_all('h1')[0].text)
```

```python
# –¥–ª—è –ø–∞—Ä—Å–∏–Ω–≥–∞ —Ç–µ–∫—Å—Ç–∞: —Å–æ–±–µ—Ä–∏—Ç–µ –≤—Å–µ —Ç–µ–∫—Å—Ç—ã –∏–∑ —Ç–µ–≥–∞ p –∏ —Å–æ–µ–¥–∏–Ω–∏—Ç–µ –≤ —Å—Ç—Ä–æ–∫—É
for x in soup.find_all('p'):
    print(x.text)
```

```python
# –∞–≤—Ç–æ—Ä–∞ —Å–æ–±—Ä–∞—Ç—å —Å–ª–æ–∂–Ω–µ–µ:
# –¥–æ—Å—Ç–∞–µ–º –∏–∑ —Ç–µ–≥–∞ meta –∫–∞–∫ –Ω–∞ –ø–∞—Ä–µ –∏–ª–∏ –∏–∑ —Ç–µ–≥–∞ a —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π –Ω–∞ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ –∞—Ç—Ä–∏–±—É—Ç–∞ rel
print(soup.find_all('meta', {'name' : 'author'})[0].attrs['content'])
```

```python
# —Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏:
# <a rel="category tag"> –Ω—É–∂–Ω—ã–µ –Ω–∞–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ </a>
# –º–æ–∂–Ω–æ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å —Ç–∞–∫:
# if i.get('rel') == ['category', 'tag']
for z in soup.find_all('a'):
  if z.get('rel') == ['category', 'tag']:
    print(z.text)
```
